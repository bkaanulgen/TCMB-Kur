```python
!pip install evds

from evds import evdsAPI
import pandas as pd
import numpy as np
evds = evdsAPI('kvlUQV03Zp')

term = '2023-07'

term_year = term[:4]
term_month = term[-2:]
term_end_day = pd.Period('{}-01-{}'.format(term_month, term_year), freq='M').end_time.day

start_date_str = '01-{}-{}'.format(term_month, term_year)
end_date_str = '{}-{}-{}'.format(term_end_day, term_month, term_year)

data_list = ['TP.DK.AUD.A.YTL', 'TP.DK.CAD.A.YTL', 'TP.DK.CHF.A.YTL', 'TP.DK.DKK.A.YTL', 'TP.DK.EUR.A.YTL', 'TP.DK.GBP.A.YTL', 'TP.DK.JPY.A.YTL', 'TP.DK.KWD.A.YTL', 'TP.DK.NOK.A.YTL', 'TP.DK.SAR.A.YTL', 'TP.DK.SEK.A.YTL', 'TP.DK.USD.A.YTL']

df_evds_avg = evds.get_data(data_list, startdate = start_date_str, enddate = end_date_str, aggregation_types='avg', frequency = 5)
df_evds_end = evds.get_data(data_list, startdate = start_date_str, enddate = end_date_str)

df_avg = pd.DataFrame()
df_avg['Döviz'] = data_list
df_avg['Döviz'] = df_avg['Döviz'].apply(lambda x: x[6:9])
df_avg[term] = list(df_evds_avg.iloc[0, 1:])
df_avg = df_avg.sort_values('Döviz').set_index('Döviz')

df_avg['Ortalama'] = np.round(df_avg[term] * 1_000_000).astype(int).astype(str)
df_avg.loc['JPY', 'Ortalama'] = '{:.2f}'.format(np.round(df_avg.loc['JPY', term] * 10_000, 2))

df_evds_end_fix = df_evds_end
df_evds_end_fix = df_evds_end_fix[df_evds_end_fix.iloc[:,1].notnull()].sort_values('Tarih', ascending = False).head(1)

df_end = pd.DataFrame()
df_end['Döviz'] = data_list
df_end['Döviz'] = df_end['Döviz'].apply(lambda x: x[6:9])
df_end[term] = list(df_evds_end_fix.iloc[0, 1:])
df_end = df_end.sort_values('Döviz').set_index('Döviz')

df_end['Yıl Sonu'] = np.round(df_end[term] * 1_000_000).astype(int).astype(str)
df_end.loc['JPY', 'Yıl Sonu'] = '{:.2f}'.format(np.round(df_end.loc['JPY', term] * 10_000, 2))
```
